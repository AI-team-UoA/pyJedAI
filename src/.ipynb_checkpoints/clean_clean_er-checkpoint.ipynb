### Data Reading

```python
d1 = pd.read_csv("../data/D2/abt.csv", sep='|', engine='python').astype(str)
d2 = pd.read_csv("../data/D2/buy.csv", sep='|', engine='python').astype(str)
gt = pd.read_csv("../data/D2/gt.csv", sep='|', engine='python')
```

```python
# d1 = d1.loc[[0,1]]
# d2 = d2.loc[[53,151]]
# gt = gt.loc[[594, 702]]
```

```python
from datamodel import Data

data = Data(
    dataset_1=d1,
    attributes_1=['id','name','description'],
    id_column_name_1='id',
    dataset_2=d2,
    attributes_2=['id','name','description'],
    id_column_name_2='id',
    ground_truth=gt,
)

data.process(cora_text_cleaning_method)
```

```python
data.print_specs()
```

### Schema Clustering

```python
# TODO valentine
```

### Block Building

```python
from blocks.building import (
    StandardBlocking,
    QGramsBlocking,
    SuffixArraysBlocking,
    ExtendedSuffixArraysBlocking,
    ExtendedQGramsBlocking
)
```

```python
blocks = StandardBlocking().build_blocks(data)
```

```python
blocks = QGramsBlocking(
    qgrams=2
).build_blocks(data)
```

```python
blocks = ExtendedQGramsBlocking(
    qgrams=2,
    threshold=0.9
).build_blocks(data)
```

```python
print_blocks(blocks, data.is_dirty_er)
```

```python
Evaluation().report(blocks, data)
```

### Block Cleaning

```python
from blocks.cleaning import (
    BlockFiltering
)
```

```python
filtered_blocks = BlockFiltering(
    ratio=0.9
).process(blocks, data)
```

```python
# print_blocks(filtered_blocks, data.is_dirty_er)
```

```python
Evaluation().report(filtered_blocks, data)
```

### Comparison Cleaning - Meta Blocking

```python
from blocks.purging import (
    ComparisonsBasedBlockPurging
)
```

```python
cleaned_blocks = ComparisonsBasedBlockPurging(
    smoothing_factor=0.008
).process(blocks, data)
```

```python
Evaluation(data).report(cleaned_blocks, data)
```

```python
from blocks.comparison_cleaning import (
    WeightedEdgePruning,
    WeightedNodePruning,
    CardinalityEdgePruning,
    CardinalityNodePruning,
    BLAST,
    ReciprocalCardinalityNodePruning,
    ReciprocalCardinalityWeightPruning,
    ComparisonPropagation
)
```

```python
%%time
candidate_pairs_blocks = WeightedEdgePruning(
    weighting_scheme='CBS'
).process(filtered_blocks, data)
```

```python
candidate_pairs_blocks = WeightedNodePruning(
    weighting_scheme='CBS'
).process(filtered_blocks, data)
```

```python
print_candidate_pairs(candidate_pairs_blocks)
```

```python
Evaluation(data).report(candidate_pairs_blocks, data)
```

### Entity Matching

```python
from matching.similarity import EntityMatching
```

```python
attr = ['author', 'title']
# or with weights
attr = {
    'Name' : 0.6, 
    'Aggregate Value' : 0.4
}

EM = EntityMatching(
    metric='jaccard', 
    similarity_threshold=0.5
    # embedings=None, # gensim
    # attributes=attr,
    # qgram=2 # for ngram metric or jaccard
)
```

```python
pairs_graph = EM.predict(filtered_blocks, data)
```

```python
%%time

attr = {
    'Name' : 0.6, 
    'Aggregate Value' : 0.4
}

EM = EntityMatching(
    metric='jaccard',
    similarity_threshold=0.5
)

pairs_graph = EM.predict(candidate_pairs_blocks, data)
```

```python
draw(pairs_graph)
```

### Entity Clustering

```python
from clustering.connected_components import ConnectedComponentsClustering
```

```python
clusters = ConnectedComponentsClustering().process(pairs_graph)
```

```python
print_clusters(clusters)
```

```python
e = Evaluation(data)

e.report(clusters)
```

```python
e.confusion_matrix()
```
